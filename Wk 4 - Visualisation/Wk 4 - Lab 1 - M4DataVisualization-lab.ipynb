Import pandas module.

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
import matplotlib.pyplot as plt

sql = '''
SELECT DatabaseDesireNextYear AS Database, COUNT(Respondent) AS Count
FROM DatabaseDesireNextYear
GROUP BY Database
ORDER BY Count DESC
LIMIT 5;
'''

df = pd.read_sql(sql, conn)

ax = df.plot(kind='pie', labels=df['Database'], y='Count', shadow=True, figsize=(10,20), autopct='%1.1f%%', title='Top 5 Most Desirable Databases to Learn Next Year')
plt.show()
```

```python
# Find out how popular python is to learn next year

sql = '''
SELECT LanguageDesireNextYear, Count(LanguageDesireNextYear) AS Count
FROM LanguageDesireNextYear
GROUP BY LanguageDesireNextYear
ORDER BY Count DESC;
'''

df = pd.read_sql(sql, conn)
df.head()
```

```python
# Find out how many respondents work with SQL

sql = '''
SELECT Count(LanguageWorkedWith) AS SQL_Count
FROM LanguageWorkedWith
WHERE LanguageWorkedWith = "SQL";
'''

df = pd.read_sql(sql, conn)
df.head()
```

```python
# Find out how many respondents work with MySQL

sql = '''
SELECT COUNT(*) AS MySQL_Only_Count 
FROM(
    SELECT Respondent, DatabaseWorkedwith
    FROM DatabaseWorkedWith d
    GROUP BY Respondent
    HAVING COUNT(RESPONDENT) = 1
    )
WHERE DatabaseWorkedWith = "MySQL";

'''

df = pd.read_sql(sql, conn)
df.head()
```

```python
# Get developer environments for responders

sql = '''
SELECT DevType, COUNT(DevType) AS Environ_Count
FROM DevType
WHERE DevType LIKE "%Developer%"
GROUP BY DevType
ORDER BY Environ_Count DESC;
'''

df = pd.read_sql(sql, conn)
df.head()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
sql = '''
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE Age BETWEEN 30 AND 35;
'''

df = pd.read_sql(sql, conn)
df = df.groupby('Age').median().reset_index()

fig, ax = plt.subplots()
ax.bar(df['Age'], df['WorkWeekHrs'], label='WorkWeekHrs')
ax.bar(df['Age'], df['CodeRevHrs'], label='CodeRevHrs')
ax.set_ylabel('Hours')
ax.set_xlabel('Age')
ax.legend()

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
sql = '''
SELECT ConvertedComp, Age
FROM master
WHERE Age BETWEEN 45 AND 60;
'''

df = pd.read_sql(sql, conn)
df = df.groupby('Age').median().reset_index()

df.plot(x='Age', y='ConvertedComp')
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
sql = '''
SELECT MainBranch, count(MainBranch) as count
FROM master
GROUP BY MainBranch;
'''

df = pd.read_sql(sql, conn)

df.plot(kind='barh')
```
